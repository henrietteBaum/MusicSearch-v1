## Button

In Tkinter werden die verschiedenen Konfigurationsoptionen (wie z.B. text, fg für Vordergrundfarbe, bg für Hintergrundfarbe, width, height und eben command) für ein Widget intern als ein Dictionary (Wörterbuch) gespeichert.

Der Zugriff über eckige Klammern [] ist in Python die Standardmethode, um auf Schlüssel (Keys) in einem Dictionary zuzugreifen oder deren Werte zu ändern. In diesem Fall ist command der Schlüssel (als String) und self.button_clicked der zuzuweisende Wert (die Funktion).

```python
# Optionen werden als benannte Argumente (keyword arguments) übergeben

self.button = tk.Button(self, text="Klick mich", command=self.button_clicked)
```

### Dictionary Syntax:

```python
self.button['command'] = self.button_clicked
```

Methode B: Später, nach der Erstellung, mit der Dictionary-Syntax
Diese Methode wird verwendet, um eine Option nach der Erstellung des Widgets zu ändern. Die Zeile, nach der Sie gefragt haben, ist ein Beispiel dafür:

```python
self.button = tk.Button(self, text="Klick mich") 
# ... später im Code ...
self.button['command'] = self.button_clicked # Hier wird die Option nachträglich gesetzt oder geändert
```

### Warum die eckigen Klammern?

Die Verwendung der eckigen Klammern ist notwendig, weil command ein String-Schlüssel ist, der in der internen Konfigurations-Dictionary des Buttons gesucht wird.

Es handelt sich nicht um den Zugriff auf ein direktes Attribut der Python-Klasse Button (wie z.B. self.button.command), sondern um den Zugriff auf einen Eintrag im Konfigurationsspeicher des zugrunde liegenden Tk-Widgets.
Zusammenfassend: ['command'] ist die Dictionary-Syntax, um auf die Konfigurationsoption "command" des Tkinter-Buttons zuzugreifen und ihr die Funktion self.button_clicked zuzuweisen, die ausgeführt werden soll, wenn der Button gedrückt wird.

```python
## Frame App:
```

```python
import tkinter as tk
from tkinter import ttk
from tkinter.messagebox import showinfo

class MainFrame(ttk.Frame):
    def __init__(self, container):
        super().__init__(container)

        options = {'padx': 10, 'pady': 20}

        # label
        self.label = ttk.Label(self, text='Hello, Tkinter!')
        self.label.pack(**options)

        # button
        self.button = ttk.Button(self, text='Click Me')
        self.button['command'] = self.button_clicked
        self.button.pack(**options)

        # show the frame on the container
        self.pack(**options)

    def button_clicked(self):
        showinfo(title='Information',
                 message='Hello, Tkinter!')

class App(tk.Tk):
    def __init__(self):
        super().__init__()
        # configure the root window
        self.title('My Awesome App')
        self.geometry('600x400')

if __name__ == "__main__":
    app = App()
    frame = MainFrame(app)
    app.mainloop()
```
